```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = np.random.randint(0,101,20,dtype='int')
print(dataset)
```

```python
plt.plot(dataset, marker='o' , color='r')
plt.show()
```

```python
datasetN=pd.Series(data=dataset)
```

```python
d=datasetN.describe()
print('mean:',datasetN.mean())
print('mod:',datasetN.mode()[0])
print('Lower Quartile:',datasetN.quantile(0.25))
print('median:',datasetN.median())
print('upper Quartile:',datasetN.quantile(0.75))
print('InterQuartile range:',datasetN.quantile(0.75)-datasetN.quantile(0.25))

s=float(0)
for i in range(0,20):
    s = s + (dataset[i] - np.mean(dataset))**2

sd = np.sqrt(s/dataset.size)
print('SD Manual:',sd)
print('SD Manual:',dataset.std())
```

```python
#part 1
dataset2=np.empty(shape=(5,10), dtype='int')
for i in range(0,5):
    for j in range(0,10):
        dataset2[i][j] = int(np.random.randint(0,9))
        
print(dataset2)
```

```python
plt.plot(dataset2[0,], marker='o' , color='g')
plt.plot(dataset2[1,], marker='s' , color='b')
plt.show()
```

```python
#part 2
print("mean by row: ",np.mean(dataset2,axis=1))
print("median by row: ",np.median(dataset2,axis=1))
print("Std by row: ",np.std(dataset2,axis=1))
```
