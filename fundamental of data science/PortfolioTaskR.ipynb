```r
# use inbuilt tidyr who data set
who_data  = tidyr::who
who_data
```

```r
#1:(2 marks): Gather together all the columns from _new_spm014 to _newrelf65

who1 <-who_data %>%
  pivot_longer(
    cols = new_sp_m014:newrel_f65,
    names_to = "key",
    values_to = "value",
    values_drop_na = TRUE
  ) 
who1
```

```r
#2:  (2 marks): Make variable names consistent....
who2 <-who1 %>% mutate(key = str_replace(key, "newrel","new_rel"))
who2
```

```r
#3:	(2 mark): Run the following code

who3 <- who2 %>% separate(key,c("new", "type", "sexage"),sep="_")
who3
```

```r
#4: (1 mark): Separate sexage into sex and age: Use the function separate(). Name the dataset who4 	
who4 <-who3 %>% separate(sexage,c("sex","Age"),sep="(?<=[a-z])(?=[0-9])")
who4
```

```r
#  Q1. Plot and explain: Which vehicle brand (or manufacturer), offers the best mpg in both city and in the highway? 


unique(mpg$manufacturer)

df_grp_manufacturer = mpg %>% group_by(manufacturer)  %>%
  summarise(city_mpg = mean(cty),
            high_waympb = mean(hwy),
            .groups = 'drop')

df_grp_manufacturer #  Group by best mpg in city and Highway

# Now we need to convert data in melt form 

df_manufacturer <- melt(df_grp_manufacturer[,c('manufacturer','city_mpg','high_waympb')],id.vars = 1)
df_manufacturer # melted data

```

```r
# Now plot data to find the brand name whose mpg best in both city and highway
ggplot(df_manufacturer,aes(x = variable,y = value)) +
  geom_bar(aes(fill = manufacturer),stat = "identity",position = "dodge") +
  ggtitle("Car brand vs best mpg in city and highway both") +  scale_y_log10()

# So our answer is in bar plot.Using Bar plot any easily visulize that Honda is the one who have best mpg in both city and highway. 
```

```r
# Q2. Plot and explain: Which type of car, regarding their displ range (size of engine) has the lowest mpg in the city
# categorised by the vehicle type (e.g., compact, suv or 2seaters defined in the variable class)?

df_type_car <- mpg %>%group_by(displ, class) %>% summarise(city_mpg = min(cty), .groups = 'drop')
df_type_car # Dataframe have lowest mpg in city using factor displ
```

```r
# Plot data
ggplot(df_type_car, aes(x=class, y=city_mpg, fill = factor(displ))) + 
  geom_bar(stat = 'identity', position = 'dodge') +ggtitle("Lowest mpg in city regarding displ")+
  geom_text(aes(label = city_mpg), vjust = -0.2, size = 5, position = position_dodge(0.9))

# So there are two type of car "pickup" and "SUV" has the lowest mpg in the city regarding displ range. 
```

```r
# Q3. Plot and explain: Which type of car, regarding their displ range (size of engine) has the best mpg performance
# in both city and highway? Display the results categorised by the number of cylinders and the vehicle type.
# You are a buyer who wants a high litre engine vehicle and drives mostly in the highway, which type of car would you choose? 

df_best_performance_car <- mpg %>%group_by(displ) %>% summarise(city_mpg = max(cty),hwy_mpg = max(hwy),  .groups = 'drop')
df_best_performance_car
```

```r
# need to melt data
melt_car_type <- melt(df_best_performance_car[,c('displ','city_mpg','hwy_mpg')],id.vars = 1)
melt_car_type
```

```r
# Plot melted data 
ggplot(melt_car_type,aes(x = displ,y = value)) +
  geom_bar(aes(fill = variable),stat = "identity",position = "dodge") +ggtitle("Best mpg in both city and highway Regarding displ")+
  scale_y_log10()
# So using barplot we can identify that car have displ value less then 2 have best mpg in both city and highwey 
```

```r
# Now we need to display results categorised by the number of cylinders and the vehicle type
df_cat_by_cyl_typ = mpg %>% group_by(class)  %>%
  summarise(cyl = mean(cyl),
            .groups = 'drop')
 
df_cat_by_cyl_typ # data categorised by the number of cylinders and the vehicle type 

# plo categorised by the number of cylinders and the vehicle type 
ggplot(df_cat_by_cyl_typ, aes(class, cyl)) +
  ggtitle("Car class categorised by Cylinder and type ")+geom_col()

# 2seater car have more cylinder then other type of car.
```

```r
# Now need to choose car who have high liter engine vehicle and drive mostly in the highway 
# so first grouped data by displ and hwy regarding Class factor.
df_grp_class = mpg %>% group_by(class)  %>%
  summarise(displ_mpg = mean(displ),
            hwy_mpg = mean(hwy),
            .groups = 'drop')

df_grp_class

# Need to melt data
melt_df_grp_class<- melt(df_grp_class[,c('class','displ_mpg', 'hwy_mpg')],id.vars = 1)
melt_df_grp_class
```
